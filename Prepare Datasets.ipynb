```python
import pandas as pd
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
from tqdm.autonotebook import tqdm, trange
import getInvalid
import PostSW
import PreSW
```

```python
#Load Batches of results files, process, generate FRAMED Dataset
interval=200
files=[]
for i in range(22):
    #Make a list of filenames
    files.append("Original frame simulation results/Batch_1/results_new_%d_%d.csv" % (i*interval, i*interval+199))  
#Make a list of bike indices
augdf=pd.read_csv("Processed_structural_data_backup.csv", index_col=0)
indices=augdf.index
origindices=indices
    
#Name the batches
batchnames=["1_"+str(x+1) for x in list(range(len(files)))]

#Call our batch data extraction function
alldata = PostSW.extract_data_batch(files, indices, batchnames, interval)


#Batch 2
files=[]
for i in range(9):
    files.append("Original frame simulation results/Batch_2/results_resim_%d_%d.csv" % (i*interval, i*interval+199))
indices=np.load("Original frame simulation results/Batch Info/batch2_backup.npy")
batchnames=["2_" + str(x+1) for x in list(range(len(files)))]
b2 = PostSW.extract_data_batch(files, indices, batchnames, interval)


#Batch 3
files=[]        
for i in range(6):
    files.append("Original frame simulation results/Batch_3/results_mat_%d_%d.csv" % (i*interval, i*interval+199))
    indices=np.load("Original frame simulation results/Batch Info/batch3.npy")
batchnames=["3_" + str(x+1) for x in list(range(len(files)))]    
mat = PostSW.extract_data_batch(files, indices, batchnames, interval)

#Merge batches 2 and 3 into batch 1 (batches 2 and 3 correct errors in some of batch 1's data)
intersection=set(b2.index).intersection(alldata.index)
alldata.drop(intersection, axis=0, inplace=True)
alldata=pd.concat([alldata, b2.loc[intersection]])

intersection=set(mat.index).intersection(alldata.index)
alldata.drop(intersection, axis=0, inplace=True)
alldata=pd.concat([alldata, mat.loc[intersection]])
alldata.sort_index(inplace=True)



#Calculate Quantities of Interest
alldata=PostSW.calcQOA(alldata)


#Rename Materials
alldata=PostSW.renameMats(alldata)


#Generate Validity data
#Result columns to discard
discard=-11

bikeddata=pd.read_csv("biked_predrop.csv", index_col=0)
invalidindex=getInvalid.getInvalid(bikeddata)[0]
invalidindex=set(invalidindex)
bikedindex=set(bikeddata.index)
# bikedindex=set([x for x in bikeddata.index if x <= alldata.index[-1]])
validindex=set(alldata.index)
validindex=validindex-invalidindex
dnsindex=bikedindex-invalidindex-validindex

# print(len(bikedindex))
# print(len(validindex))
# print(len(invalidindex))
# print(len(dnsindex))
alldata=alldata.loc[validindex]

bikeddata.columns=[a[1:] for a in bikeddata.columns] #Remove Leading space from each column title
bikeddata=PostSW.renameMats(bikeddata)

valid=bikeddata.loc[validindex]
# valid=valid[alldata.columns[:discard]]
valid.loc[:, "valid"]=np.zeros(len(valid.index))

invalid=bikeddata.loc[invalidindex]
# invalid=invalid[alldata.columns[:-11]]
invalid.loc[:, "valid"]=np.ones(len(invalid.index))

dns=bikeddata.loc[dnsindex]
# invalid=invalid[alldata.columns[:-11]]
dns.loc[:, "valid"]=np.ones(len(dns.index))*2

validdf=pd.concat([valid,invalid,dns])
validdf.sort_index(axis=0, inplace=True)

#Set bridge booleans
alldata=PostSW.setBridges(alldata)
validdf=PostSW.setBridges(validdf)

invalid_bikes, perc=getInvalid.getInvalid(alldata)
alldata.drop(invalid_bikes, axis=0, inplace=True)
validdf.loc[invalid_bikes, "valid"]=0

validdf_oh=PostSW.oneHot(validdf)
# print(validdf)
validdf_oh.to_csv("validity.csv")   



alldata.to_csv("all_structural_data.csv")  

# alldata_aug=DataAugment.featureAugment(alldata)
# alldata_aug.to_csv("all_structural_data_aug.csv") 

alldata_oh=PostSW.oneHot(alldata)

# print(alldata)
alldata_oh.to_csv("all_structural_data_oh.csv")  
# alldata_oh_aug=DataAugment.featureAugment(alldata_oh)
# alldata_oh_aug.to_csv("all_structural_data_oh_aug.csv") 
```
